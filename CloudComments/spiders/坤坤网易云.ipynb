```python
import pandas as pd
from pymongo import MongoClient
client = MongoClient()
collection = client['text']['comments2']
data = list(collection.find())
df = pd.DataFrame(data)
df.head(20)
```

```python
df.info()
```

```python
df = df.drop_duplicates(['id'])
```

```python
df.info()
```

```python
with open('comments2.txt','ab') as f:
    for comment in df['comment']:
        f.write(comment.encode('utf-8'))
        f.write(b'\n')
    print('写入完毕')
```

```python
import jieba
txt = open('comments2.txt','r',encoding='utf-8').read()
words = jieba.lcut(txt)
print(type(words))
print(len(words))
#print(words[:10])
word_counts = {}
for word in words:
    word_counts[word] = word_counts.get(word,0) + 1
excludes = ['。','，','：',',','\n','的','了','我','是','看','都','！','这','也','在','听','你','有','人','很','想', '啊','不',
           '就','…','？','着','和','会','又','这个','一个','吗','没有','这部','完','到','说','现在','让','看到','吧','还',
            '真的','时候','电视剧','没','每次','知道','还是','给','一起','里','就是','他','自己','你们','不是','什么','呢','怎么',
            '哦','我们','去','来','那','被','别人','要','“','能','对','多','”','觉得','她','这么','太'
           ]
for word in excludes:
    if word in word_counts:
        del word_counts[word]
items = list(word_counts.items())
items.sort(key=lambda x:x[1],reverse=True)
items[:20]
```

```python
from pyecharts import options as opts
from pyecharts.charts import Page, WordCloud
word = WordCloud()
word.add('',items[:20],word_size_range=[20, 100])
word.set_global_opts(title_opts=opts.TitleOpts(title="WordCloud展示"))
word.render_notebook()
```
